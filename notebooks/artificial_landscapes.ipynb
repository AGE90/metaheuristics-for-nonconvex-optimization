## Rastrigin function

On an $n$-dimensional domain the Rastrigin function it is defined by:

$$f(\mathbf{x}) = An + \sum_{i=1}^{n} \left[x_{i}^{2} - A\cos{2\pi x_{i}}\right]$$

where $A = 10$ and $x_{i} \in [-5.12, 5.12]$. It has a global minimum at $\mathbf{x} = \mathbf{0}$ where $f(\mathbf{x}) = 0$. 

```python
x1 = np.linspace(-5.12, 5.12, 1000)     
x2 = np.linspace(-5.12, 5.12, 1000)
X1, X2 = np.meshgrid(x1, x2)

Y = rastrigin(X1, X2)
```

```python
fig, ax = plt.subplots(figsize=(6, 4), 
                       subplot_kw={"projection": "3d"})

surf = ax.plot_surface(X1, X2, Y, 
                       cmap=cm.viridis)
cset = ax.contourf(X1, X2, Y, 
                   zdir='z', offset=np.min(Y)-30, 
                   cmap=cm.viridis, alpha=0.8)

ax.set_zlim(np.min(Y)-30, np.max(Y))

ax.view_init(25, 35)

ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$y$')
plt.tight_layout()
plt.show()
```

## Ackley function

On a 2-dimensional domain it is defined by:

$$f(x, y)=-20\exp \left[-0.2\sqrt{0.5(x^{2} + y^{2})}\right] - \exp\left[0.5\left(\cos{2\pi x} + \cos{2\pi y}\right)\right] + e + 20,$$

with its global optimum point is

$$f(0, 0)=0.$$

```python
x1 = np.linspace(-5, 5, 1000)     
x2 = np.linspace(-5, 5, 1000)
X1, X2 = np.meshgrid(x1, x2)

Y = ackley(X1, X2)
```

```python
fig, ax = plt.subplots(figsize=(6, 4), 
                       subplot_kw={"projection": "3d"})

surf = ax.plot_surface(X1, X2, Y, 
                       cmap=cm.viridis)
cset = ax.contourf(X1, X2, Y, 
                   zdir='z', offset=np.min(Y)-10, 
                   cmap=cm.viridis, alpha=0.8)

ax.set_zlim(np.min(Y)-10, np.max(Y))

ax.view_init(25, 35)

ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$y$')
plt.tight_layout()
plt.show()
```

## Rosenbrock function

On a 2-dimensional domain it is defined by:

$$f(\mathbf{x}) = \sum_{i=1}^{N-1}[100(x_{i+1}-x_{i}^{2})^{2}+(1-x_{i})^{2}].$$

where ${x}=(x_{1},\ldots ,x_{N}) \in \mathbb{R}^{N}$.


```python
x1 = np.linspace(-2, 2, 1000)     
x2 = np.linspace(-1, 3, 1000)
X1, X2 = np.meshgrid(x1, x2)

Y = rosenbrock(X1, X2)
```

```python
fig, ax = plt.subplots(figsize=(6, 4), 
                       subplot_kw={"projection": "3d"})

surf = ax.plot_surface(X1, X2, Y, 
                       cmap=cm.viridis)
cset = ax.contourf(X1, X2, Y, 
                   zdir='z', offset=np.min(Y)-5e3, 
                   cmap=cm.viridis, alpha=0.8)

ax.set_zlim(np.min(Y)-5e3, np.max(Y))

ax.view_init(25, 55)

ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$y$')
plt.tight_layout()
plt.show()
```
